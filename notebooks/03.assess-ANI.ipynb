```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

sns.set_context("paper")
import matplotlib.ticker as ticker

from tqdm import tqdm # progress bars :)
tqdm.pandas()
```

```python
ep = pd.read_csv("gtdb-rs202.evolpaths.merged.csv.gz")
ep
```

```python
ep.columns
```

```python
dna_only = ep[ep["alphabet"] == "nucleotide"]
dna_only = dna_only[dna_only["pyani_ident"].notna()] # only grab comparisons we can work with
dna_only
```

```python
def ani_in_ci(row):
    row["ani_in_mc_ci"] = False
    row["ani_in_jaccard_ci"] = False
    ani = float(row["pyani_ident"])
    mc_low = float(row["mcANI_low"])
    mc_high = float(row["mcANI_high"])
    if mc_low <= ani <= mc_high:
        row["ani_in_mc_ci"] = True
    jani_low = float(row["jani_low"])
    jani_high = float(row["jani_high"])
    if jani_low <= ani <= jani_high:
        row["ani_in_jaccard_ci"] = True
    return row
```

```python
dna_only = dna_only.progress_apply(ani_in_ci, axis=1)
dna_only
```

```python
in_mcCI = dna_only[dna_only["ani_in_mc_ci"] ==True]
in_mcCI
```

```python
in_mcCI.groupby(["scaled", "ksize"])[["ani_in_mc_ci"]].describe()
```

```python
in_jCI = dna_only[dna_only["ani_in_jaccard_ci"] ==True]
in_jCI
```

```python
in_jCI.groupby(["scaled", "ksize"])[["ani_in_jaccard_ci"]].describe()
```
