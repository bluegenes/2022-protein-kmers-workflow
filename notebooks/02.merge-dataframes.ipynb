```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
pathcompare = pd.read_csv("gtdb-rs202.pathcompare.csv.gz")
fastani = pd.read_csv("gtdb-rs202.path-fastani.csv.gz")
pyani_ANIm = pd.read_csv("gtdb-rs202.pyani-ANIm.csv.gz")
pyani_ANIb = pd.read_csv("gtdb-rs202.pyani-ANIb.csv.gz")
compareM = pd.read_csv("gtdb-rs202.path-compareM.proteome.csv.gz")
```

```python
# replace zeroes with np.nan
cols = ["jaccard_ani", "containA_ani", "containBani", "mcANI"]
pathcompare[cols] = pathcompare[cols].replace({0:np.nan})
pathcompare[cols] = pathcompare[cols].replace({1:np.nan})
```

```python
pathcompare
```

```python
fastani = fastani.drop_duplicates()
pyani_ANIm = pyani_ANIm.drop_duplicates()
pyani_ANIb = pyani_ANIb.drop_duplicates()
compareM = compareM.drop_duplicates()
```

```python
#pyani_cols = ['pyani_ident', 'pyani_coverage', 'pyani_aln_length', 'pyani_sim_errors', 'pyani_hadamard']
ANIb_rename = {'pyani_ident': 'ANIb_ident', 
                'pyani_coverage':'ANIb_coverage', 
                'pyani_aln_length': "ANIb_aln_length", 
                "pyani_sim_errors": "ANIb_sim_errors", 
                "pyani_hadamard": "ANIb_hadamard" }

pyani_cols = list(ANIb_rename.values())
print(pyani_cols)

pyani_ANIb.rename(columns=ANIb_rename, inplace=True)
pyani_ANIb[pyani_cols] = pyani_ANIb[pyani_cols].replace({0:np.nan, 1:np.nan})
# filter pyANI -- >=30% coverage and >=70% identity
pyani_ANIb = pyani_ANIb[(pyani_ANIb["ANIb_coverage"] >= 0.3) & (pyani_ANIb["ANIb_ident"] >= 0.7)]
pyani_ANIb
```

```python
ANIm_rename = {'pyani_ident': 'ANIm_ident', 
                'pyani_coverage':'ANIm_coverage', 
                'pyani_aln_length': "ANIm_aln_length", 
                "pyani_sim_errors": "ANIm_sim_errors", 
                "pyani_hadamard": "ANIm_hadamard" }

pyani_cols = list(ANIm_rename.values())

pyani_ANIm.rename(columns=ANIm_rename, inplace=True)
pyani_ANIm[pyani_cols] = pyani_ANIm[pyani_cols].replace({0:np.nan, 1:np.nan})
# filter pyANI -- >=30% coverage and >=70% identity
pyani_ANIm = pyani_ANIm[(pyani_ANIm["ANIm_coverage"] >= 0.3) & (pyani_ANIm["ANIm_ident"] >= 0.7)]
pyani_ANIm
```

```python
# merge pyANI
pyani = pd.merge(pyani_ANIb, pyani_ANIm, how='outer',on = ["comparison_name", "anchor_name", "compare_name", "path", "lowest_common_rank"])#,indicator=True
pyani
```

```python
compareM.rename(columns={"mean_aai": "compareM_mean_aai", "std_aai": "compareM_std_aai", "ref_name": "compare_name"}, inplace=True)
fastani.rename(columns={"ref_name": "compare_name"}, inplace=True)
# get sp same units
compareM["compareM_mean_aai"] = compareM["compareM_mean_aai"]/100
compareM["compareM_std_aai"] = compareM["compareM_std_aai"]/100
fastani["fastani_ident"] = fastani["fastani_ident"]/100
```

```python
# filter CompareM AAI -- >=4% orthologous fraction and >=40% identity
compareM = compareM[(compareM["orthologous_fraction"] >= 4) & (compareM["compareM_mean_aai"] >= 0.4)]
compareM
```

```python
# merge outside tools first
other_compare = pd.merge(compareM, pyani, how="outer", on = ["comparison_name", "anchor_name", "compare_name", "path", "lowest_common_rank"])#,indicator=True
other_compare
```

```python
# merge outside tools first
other_compare = pd.merge(other_compare, fastani, how="outer", on = ["comparison_name", "anchor_name", "compare_name", "path", "lowest_common_rank"])#,indicator=True
other_compare
```

```python
other_compare.columns
```

```python
all_merged = pd.merge(pathcompare, other_compare, how="left", on = ["comparison_name", "anchor_name", "compare_name", "path", "lowest_common_rank"])
all_merged["min_containment_ani"] = all_merged[["containA_ani", "containBani"]].min(axis=1)
all_merged["avg_containment_ani"] = all_merged[["containA_ani", "containBani"]].mean(axis=1)
all_merged
```

```python
all_merged.columns
```

```python
prot_vals_to_replace= {"jaccard_ani": "jaccard AAI", 
                  "mcANI": "maxContain AAI",
                  "containA_ani": "containA AAI", 
                  "containBani": "containB AAI",
                  "min_containment_ani": "minContain AAI",
                  "avg_containment_ani": "avgContain AAI",
                  "compareM_mean_aai": "compareM AAI",
                  "ANIb_ident": "ANIb ident", 
                  "ANIm_ident": "ANIm ident", 
                  "fastani_ident": "fastani ANI"}

nucl_vals_to_replace= {"jaccard_ani": "jaccard ANI", 
                  "mcANI": "maxContain ANI", 
                  "containA_ani": "containA ANI", 
                  "containBani": "containB ANI", 
                  "fastani_ident": "fastani ANI", 
                  "min_containment_ani": "minContain ANI",
                  "avg_containment_ani": "avgContain ANI",
                  "compareM_mean_aai": "compareM AAI",
                   "ANIb_ident": "ANIb ident", 
                  "ANIm_ident": "ANIm ident"}
```

```python
dna_only = all_merged[all_merged["alphabet"] == "nucleotide"]
dna_only.rename(columns=nucl_vals_to_replace, inplace=True)
dna_only
```

```python
prot_only = all_merged[all_merged["alphabet"] == "protein"]
prot_only.rename(columns=prot_vals_to_replace, inplace=True)
prot_only
```

```python
dna_only.columns
```

```python
all_merged.to_csv("gtdb-rs202.evolpaths.all.csv.gz", index=False)
dna_only.to_csv("gtdb-rs202.evolpaths.nucleotide.csv.gz", index=False)
prot_only.to_csv("gtdb-rs202.evolpaths.protein.csv.gz", index=False)
```
